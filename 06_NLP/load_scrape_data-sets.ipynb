#### PDF's

Bsp. um sich alle PDF's einer Webseite zu ziehen in UNIX: 

`wget -r -P ./pdfs -A pdf https://www.christian-lindner.de/reden`

`for i in *; do pdftotext "$i"; done`

```python
# ...oder in python direkt:
import os
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup

url = "https://www.christian-lindner.de/reden"

#If there is no such folder, the script will create one automatically
folder_location = r'./lindner-talks'
if not os.path.exists(folder_location):os.mkdir(folder_location)

response = requests.get(url)
soup= BeautifulSoup(response.text, "html.parser")     
for link in soup.select("a[href$='.pdf']"):
    #Name the pdf files using the last portion of each link which are unique in this case
    filename = os.path.join(folder_location,link['href'].split('/')[-1])
    with open(filename, 'wb') as f:
        f.write(requests.get(urljoin(url,link['href'])).content)
```

### im html-format

```python
url = "https://theorieblog.attac.de/quo-vadis-homo-spiens/"
html = request.urlopen(url).read().decode('utf8')
html[:60]
```

```python
from bs4 import BeautifulSoup
import nltk
raw = BeautifulSoup(html, 'html.parser').get_text()
tokens = nltk.word_tokenize(raw)
print(tokens)
```

```python
tokens2 = tokens[3025:]
print(tokens2)
```

```python
text = nltk.Text(tokens2)
print(text)
type(text)
```

```python
text.concordance('Rede')
```
